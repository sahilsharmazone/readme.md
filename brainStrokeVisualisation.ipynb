```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
db=pd.read_csv("full_data.csv")
db.head()
```

```python
db.dropna(axis=0, inplace=True)
db.info()
```

```python
corr = db[db.columns].corr()
sns.heatmap(corr, cmap="hsv", annot = False)     
plt.title('Heatmap for Correlation of Parameters')
sns.set(rc={'figure.figsize':(13, 10)})
plt.tight_layout()
plt.show()
```

```python
db.hist(figsize=(17,12))
plt.show()
```

```python
db.plot(kind='density', subplots=True, layout=(10,1), sharex=False)
plt.show()
```

```python
db.plot(kind='box', subplots=True, layout=(10,1), sharex=False,sharey=False)
plt.show()
```
